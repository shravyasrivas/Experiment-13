```python
#preprocessing on a dataset-1
import pandas as pd #debug dataset to give to ml algorith

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
df
#to use pandas we need tabular form
#data cleaning - NaN in place of missing values here -
```

```python
#df.replace("-",np.nan,inplace=True)
#df
#covert columns to numeric ehere meeded
df["Age"]=pd.to_numeric(df["Age"],errors='coerce')
df["Marks"]=pd.to_numeric(df["Marks"],errors='coerce')
df
```

```python
#fill missing age with mean and fill missing marks with median
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
#making all department in capital letters fixing inconsistent text formatiing
df["Department"]=df["Department"].str.upper()#for lower right lower
df
```

```python
#convert admission date to standard datetime format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"],format="mixed",dayfirst=True)
df
#NaT-not a time when the date isnt valid ed 32.12.2012
#now data is clean
```
